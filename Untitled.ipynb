# Soccer Player Performance Analysis: FIFA 2022 Player Stats

## Objective
- **Goal**: Analyze performance of selected soccer players from the FIFA 2022 dataset to track key metrics, provide insights, and design a tracking system.
- **Scope**: Focus on 6 players' stats (overall rating, potential, shooting, passing) from the `players_22.csv` file: L. Messi, K. Mbappé, Neymar Jr, Cristiano Ronaldo, O. Giroud, and G. Bale.
- **Steps**:
  1. Load and clean the `players_22.csv` dataset, selecting relevant columns.
  2. Analyze key performance metrics (overall rating, potential growth, contribution score).
  3. Develop insights and visualize trends in Jupyter Notebook.
  4. Export data for visualizations in Power BI.
  5. Propose a tracking system for long-term monitoring.

## Load and clean dataset

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv(r'C:\Projectss\Data Analysis\fifa_2022\players_22.csv', low_memory= False)
df.head(1)
```

```python
df.columns                     
```

```python
# Drop irrelevant columns
```

```python
df=df.drop(['player_url','player_face_url', 'club_logo_url',
       'club_flag_url', 'nation_logo_url', 'nation_flag_url'],axis=1)
```

```python
df.head(1)
```

```python
# Filter for required players
players = ['L. Messi','K. Mbappé', 'Neymar Jr','Cristiano Ronaldo','O. Giroud', 'G. Bale' ]
df_players = df[df['short_name'].isin (players)]
```

```python
df_players.head()
```

```python
# Checking null values
df_players.isnull().sum()
```

```python
df_players.columns
```

```python
# Select relevant columns
df_players = df_players[['short_name', 'overall', 'potential', 'value_eur', 'wage_eur','shooting', 'passing']]
```

```python
df_players.isnull().sum()
```

```python
# Calculate potential growth (potential - overall)
df_players['potential_growth'] = df_players['potential'] - df_players['overall']
```

```python
# Create a contribution score (proxy for goals + assists, using shooting and passing)
df_players['contribution_score'] = df_players['shooting'] + df_players['passing']
```

```python
# Display the metrics
metrics = df_players[['short_name', 'overall', 'potential_growth', 'contribution_score', 'value_eur', 'wage_eur']]
```

```python
# Visualization: Bar plot for overall rating
plt.bar(metrics['short_name'], metrics['overall'], color='skyblue')
plt.xlabel('Player')
plt.ylabel('Oveerall Rating')
plt.title('Overall Rating of Selected Players')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization: Bar plot for contribution score
plt.bar(metrics['short_name'], metrics['contribution_score'], color='lightgreen')
plt.xlabel('Player')
plt.ylabel('Contribution Score (Shooting + Passing)')
plt.title('Contribution Score of Selected Players')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization: Bar plot for potential growth
plt.bar(metrics['short_name'],metrics['potential_growth'], color='salmon')
plt.xlabel('Player')
plt.ylabel('Potential Growth')
plt.title('Potential Growth of Selected Players')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization: Bar plot for market value (value_eur)
plt.bar(metrics['short_name'], metrics['value_eur'], color='gold')
plt.xlabel('Player')
plt.ylabel('Market Value (EUR)')
plt.title('Market Value of Selected Players')
plt.xticks(rotation=45)
plt.show()
```

## Updated Insights
- **L. Messi**: Likely the highest overall rating and contribution score, reflecting his all-around skill as a playmaker and scorer.
- **K. Mbappé**: High market value, indicating his status as a rising star with significant commercial appeal.
- **Neymar Jr**: Strong contribution score, driven by high passing ability, showcasing his creativity.
- **Cristiano Ronaldo**: High overall rating and shooting, reflecting his goal-scoring prowess, though market value may be lower due to career stage.
- **O. Giroud**: Likely lower overall and contribution score but still valuable as a target man with decent shooting.
- **G. Bale**: Moderate overall and contribution score, with market value reflecting his experience and past achievements.
